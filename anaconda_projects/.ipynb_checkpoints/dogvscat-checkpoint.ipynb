```python
import os
import cv2
import numpy as np
import matplotlib.pylab as plt
from sklearn.svm import SVC
import os
import cv2
import numpy as np

def load_images(folder, limit=2500):
    images, labels = [], []

    # half cats, half dogs
    half = limit // 2

    cat_files = [f for f in os.listdir(folder) if "cat" in f.lower()][:half]
    dog_files = [f for f in os.listdir(folder) if "dog" in f.lower()][:half]

    for filename in cat_files:
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (64, 64))
            images.append(img.flatten())
            labels.append(0)

    for filename in dog_files:
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (64, 64))
            images.append(img.flatten())
            labels.append(1)

    return np.array(images), np.array(labels)
```

```python
X, Y = load_images("dataset/train", limit=2500)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("Unique labels:", np.unique(Y, return_counts=True))
```

```python
print("Dataset shape:", X.shape, Y.shape)
```

```python
def show_images(img, labels, no_of_images=10):
    plt.figure(figsize=(15, 5))
    for i in range(no_of_images):
        plt.subplot(1, no_of_images, i + 1)
        plt.imshow(img[i].reshape(64, 64), cmap='gray')
        plt.title('Dog' if labels[i] == 1 else 'Cat')
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
show_images(X,Y,7)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
print("Unique values in Y (full dataset):", np.unique(Y, return_counts=True))



```

```python
# Train-test split
# ----------------------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
```

```python

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
print("Unique values in Y_train:", np.unique(Y_train, return_counts=True))
```

```python
# Train SVM
# ----------------------------
model = SVC(C=1, kernel='poly', gamma='auto')
model.fit(X_train, Y_train)
```

```python
import numpy as np

print("Unique values in Y_train:", np.unique(Y_train))
print("Counts in Y_train:", np.bincount(Y_train))
```

```python
train_data_prediction=model.predict(X_train)
train_accuracy_score=accuracy_score(Y_train,train_data_prediction)
print("Accuracy Score:" ,train_accuracy_score)
```

```python
test_data_prediction=model.predict(X_test)
test_accuracy_score=accuracy_score(Y_test,test_data_prediction)
print("Accuracy Score of Test Data:" ,test_accuracy_score)
```

```python
import random
```

```python

def show_predictions(X_test, Y_test, model, num_images=10):
    plt.figure(figsize=(15, 6))
    
    # Pick random indices
    indices = random.sample(range(len(X_test)), num_images)
    
    for i, idx in enumerate(indices):
        plt.subplot(2, num_images//2, i+1)
        
        # Predict
        pred = model.predict([X_test[idx]])[0]
        
        # Show image
        plt.imshow(X_test[idx].reshape(64, 64), cmap='gray')
        
        # Title
        pred_label = "Dog" if pred == 1 else "Cat"
        true_label = "Dog" if Y_test[idx] == 1 else "Cat"
        color = "green" if pred == Y_test[idx] else "red"  # green if correct, red if wrong
        plt.title(f"Pred: {pred_label}\nTrue: {true_label}", color=color, fontsize=10)
    
    plt.show()

# Example usage:
show_predictions(X_test, Y_test, model, num_images=10)
```
